Duration Analysis

```python
movies = df[df['type']=='Movie']
movies['duration'].value_counts().head()
```

Correlation HeatMap

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration_int'] = movies['duration'].str.replace(' min', '', regex=False)
movies['duration_int'] = pd.to_numeric(movies['duration_int'], errors='coerce')
```

```python
corr = movies[['release_year','duration_int']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x=movies['duration_int'])
plt.title("Movie Duration Outliers")
plt.show()
```
